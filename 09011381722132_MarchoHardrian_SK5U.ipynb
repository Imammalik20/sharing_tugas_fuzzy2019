# Tingkat Kepuasan Konsumen Berdasarkan Pelayanan, Harga, dan Kualitas Makanan Menggunakan Fuzzy Mamdani

```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt

# The universe of variables and membership functions
pelayanan = ctrl.Antecedent(np.arange(1, 10, 1), 'pelayanan')
Pelayanan_x = np.arange(0,10,1) #untuk plotting saja
kualitas = ctrl.Antecedent(np.arange(1, 10, 1), 'kualitas')
Kualitas_x = np.arange(0,10,1) #untuk plotting saja
kepuasan = ctrl.Consequent(np.arange(1, 10, 1), 'kepuasan')
Kepuasan_x = np.arange(0,10,1) #untuk plotting saja
```

Variabel Input dan Output yang digunakan :
1. Input terbagi menjadi 4 yaitu : 
Pelayanan :
 - Sangat tidak baik (nilai 1 - 4)
 - Tidak baik (nilai 3 - 6)
 - Cukup baik (nilai 5 - 8)
 - Baik (nilai 7 - 10)
 
 
Kualitas :
 - Sangat tidak berkualitas (nilai 1 - 4)
 - Tidak berkualitas (nilai 3 - 6)
 - Cukup berkualitas (nilai 5 - 8)
 - Berkualitas (nilai 7 - 10)
  
2. Output :

Kepuasaan :
 - Sangat tidak puas (nilai 1 - 4)
 - Tidak puas (nilai 3 - 6)
 - Cukup puas (nilai 5 - 8)
 - Puas (nilai 7 - 10)

Pada tahap inferencing, membuat aturan-aturan fuzzy yang dinyatakan dalam bentuk IF .... THEN ....

1.	If (Pelayanan is SANGAT TIDAK BAIK ) and (Harga is SANGAT MAHAL) and (Kualitas Makanan is SANGAT TIDAK BERKUALITAS) then (kepuasan is CUKUP PUAS)
2.	If (Pelayanan is SANGAT TIDAK BAIK ) and (Harga is MAHAL) and (Kualitas Makanan is TIDAK BERKUALITAS) then (Kepuasan is TIDAK PUAS)
3.	If (Pelayanan is SANGAT TIDAK BAIK ) and (Harga is CUKUP MURAH) and (Kualitas Makanan is CUKUP BERKUALITAS) then (Kepuasan is TIDAK PUAS)
4.	If (Pelayanan is SANGAT TIDAK BAIK ) and (Harga is MURAH) and (Kualitas Makanan is BERKUALITAS) then (Kepuasan is PUAS) dan seterusnya

# Rule Base

```python
rule1 = ctrl.Rule(pelayanan['sangat tidak baik'] & kualitas['sangat tidak berkualitas'], kepuasan['cukup puas'])
rule1.view()
rule2 = ctrl.Rule(pelayanan['sangat tidak baik'] & kualitas['tidak berkualitas'], kepuasan['tidak puas'])
rule2.view()
rule3 = ctrl.Rule(pelayanan['sangat tidak baik'] & kualitas['cukup berkualitas'], kepuasan['sangat tidak puas'])
rule3.view()
rule4 = ctrl.Rule(pelayanan['sangat tidak baik'] & kualitas['berkualitas'], kepuasan ['puas'])
rule4.view()

rule5 = ctrl.Rule(pelayanan['tidak baik'] & kualitas['sangat tidak berkualitas'], kepuasan['sangat tidak puas'])
rule5.view()
rule6 = ctrl.Rule(pelayanan['tidak baik'] & kualitas['tidak berkualitas'], kepuasan ['puas'])
rule6.view()
rule7 = ctrl.Rule(pelayanan['tidak baik'] & kualitas['cukup berkualitas'], kepuasan['cukup puas'])
rule7.view()
rule8 = ctrl.Rule(pelayanan['tidak baik'] & kualitas['berkualitas'], kepuasan['tidak puas'])
rule8.view()

rule9 = ctrl.Rule(pelayanan['cukup baik'] & kualitas['sangat tidak berkualitas'], kepuasan['sangat tidak puas'])
rule9.view()
rule10 = ctrl.Rule(pelayanan['cukup baik'] & kualitas['tidak berkualitas'], kepuasan['cukup puas'])
rule10.view()
rule11 = ctrl.Rule(pelayanan['cukup baik'] & kualitas['cukup berkualitas'], kepuasan['tidak puas'])
rule11.view()
rule12 = ctrl.Rule(pelayanan['cukup baik'] & kualitas['berkualitas'], kepuasan['puas'])
rule12.view()

rule13 = ctrl.Rule(pelayanan['baik'] & kualitas['sangat tidak berkualitas'], kepuasan['puas'])
rule13.view()
rule14 = ctrl.Rule(pelayanan['baik'] & kualitas['tidak berkualitas'], kepuasan['cukup puas'])
rule14.view()
rule15 = ctrl.Rule(pelayanan['baik'] & kualitas['cukup berkualitas'], kepuasan['sangat tidak puas'])
rule15.view()
rule16 = ctrl.Rule(pelayanan['baik'] & kualitas['berkualitas'], kepuasan['tidak puas'])
rule16.view()
```

# Rule Evaluation

```python
kepuasan_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9, rule10, rule11, rule12, rule13, rule14, rule15, rule16])
```

```python
kepuasan_ctrl.view()
```

# DEFUZZIFIKASI

```python
kepuasan = ctrl.ControlSystemSimulation(kepuasan_ctrl)
```

Input pelayanan kita masukan 5, dan kualitas 5

```python
kepuasan.input['pelayanan'] = 5
kepuasan.input['kualitas'] = 5
```

Melakukan perhitungan defuzzifikasi

```python
kepuasan.compute()
```
