#### 3. Nilai Linguistik dan Ranah Nilai Linguistik
a. Nilai Linguistik pada Variabel Nilai UAN ada 3, yaitu :
- Nilai UAN Rendah
$$ Nilai UAN Rendah [x] = 
\begin{cases}
    1 & ; x < 25  \\ 
    \frac{(50-x)}{(50-10)} & ; 25 <  x < 50 \\
    0 & ; x > 50
\end{cases}
$$

```python
uan_rendah       = fuzz.trapmf(x_nilai_uan, [0, 0, 25, 50,])

fig, (ax0) = plt.subplots(nrows= 1 ,facecolor='white', figsize=(10, 5))
ax0.plot(x_nilai_uan, uan_rendah, 'red', linewidth=1.5, label='Rendah')
ax0.set_title('Nilai UAN Rendah')
ax0.grid()
ax0.legend()
```

- Nilai UAN Sedang
$$ Nilai UAN Sedang [x] = 
\begin{cases}
    0 & ; x < 37 , x > 72 \\
    \frac{(x-37)}{(55-37)} & ; 37 <  x < 55 \\
    \frac{(72-x)}{(72-55)} & ; 55 <  x < 72
\end{cases}
$$

```python
uan_sedang       = fuzz.trimf(x_nilai_uan, [37, 55, 72])

fig, (ax0) = plt.subplots(nrows= 1 ,facecolor='white', figsize=(10, 5))

ax0.plot(x_nilai_uan, uan_sedang, 'green', linewidth=1.5, label='Sedang')
ax0.set_title('Nilai UAN Sedang')
ax0.grid()
ax0.legend()
```

- Nilai UAN Tinggi
$$ Nilai UAN Tinggi [x] = 
\begin{cases}
    0 & ; x < 62  \\ 
    \frac{(x-62)}{(87-62)} & ; 62 <  x < 87 \\
    1 & ; x > 87
\end{cases}
$$

```python
uan_tinggi       = fuzz.trapmf(x_nilai_uan, [62, 87, 100, 100])

fig, (ax0) = plt.subplots(nrows= 1 ,facecolor='white', figsize=(10, 5))

ax0.plot(x_nilai_uan, uan_tinggi, 'blue', linewidth=1.5, label='Tinggi')
ax0.set_title('Nilai UAN Tinggi')
ax0.grid()
ax0.legend()
```

- Grafik Nilai UAN

```python
fig, (ax0) = plt.subplots(nrows= 1 ,facecolor='white', figsize=(10, 5))

ax0.plot(x_nilai_uan, uan_rendah, 'red', linewidth=1.5, label='Rendah')
ax0.plot(x_nilai_uan, uan_sedang, 'green', linewidth=1.5, label='Sedang')
ax0.plot(x_nilai_uan, uan_tinggi, 'blue', linewidth=1.5, label='Tinggi')
ax0.set_title('Nilai UAN Rendah')
ax0.grid()
ax0.legend()
```
