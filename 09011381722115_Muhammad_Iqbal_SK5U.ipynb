# Penerapan Logika Fuzzy Sugeno Untuk Menentukan Jumlah Produksi Roti

Perencanaan jumlah produk dalam sebuah Pabrik Roti sangatlah penting. supaya dapat memenuhi permintaan pasar dengan tepat dan dengan jumlah yang sesuai. Faktor-faktor yang perlu diperhatikan dalam menentukan jumlah produk, antara lain adalah jumlah persediaan dan jumlah permintaan. Penulisan dan pembahasan pada penelitian ini adalah tentang sistem inferensi Fuzzy Metode Sugeno, penerapan sistem inferensi Fuzzy Metode Sugeno untuk menentukan jumlah produksi berdasarkan jumlah permintaaan dan data persediaan yang dimana data dari penulisan ini didapat dari Pabrik Roti dengan menggunakan Python.

# Proses dengan Python

# Sistem Inferensi - Bagian Rule Base

Keputusan dari implementasi fuzzy berasal dari rule-rule yang ada pada basis data. Rule yang digunakan pada kasus ini untuk mengetahui Produksi sebanyak 9 Rule dengan 3 Variabel Output : (Rendah, Sedang, Pada sistem ini berdasarkan jumlah fuzzifikasi dari variabel Permintaan yang berjumlah 3 dan fuzzifikasi dari variabel Persediaan yang berjumlah 3, maka rule yang dibuat berjumlah 3 x 3 = 9 rule

Rules :

    1) IF Permintaan=Rendah dan Persediaan=Rendah Then Produksi=Rendah.
    2) IF Permintaan=Rendah dan Persediaan=Sedang Then Produksi=Sedang.
    3) IF Permintaan=Rendah dan Persediaan=Tinggi Then Produksi=Sedang.
    4) IF Permintaan=Sedang dan Persediaan=Rendah Then Produksi=Sedang.
    5) IF Permintaan=Sedang dan Persediaan=Sedang Then Produksi=Sedang.
    6) IF Permintaan=Sedang dan Persediaan=Tinggi Then Produksi=Tinggi.
    7) IF Permintaan=Tinggi dan Persediaan=Rendah Then Produksi=Sedang.
    8) IF Permintaan=Tinggi dan Persediaan=Sedang Then Produksi=Tinggi.
    9) IF Permintaan=Tinggi dan Persediaan=Tinggi Then Produksi=Tinggi.

```python
rule1 = ctrl.Rule(permintaan['rendah'] & persediaan['rendah'], produksi['rendah'])
rule2 = ctrl.Rule(permintaan['rendah'] & persediaan['sedang'], produksi['sedang'])
rule3 = ctrl.Rule(permintaan['rendah'] & persediaan['tinggi'], produksi['sedang'])
rule4 = ctrl.Rule(permintaan['sedang'] & persediaan['rendah'], produksi['sedang'])
rule5 = ctrl.Rule(permintaan['sedang'] & persediaan['sedang'], produksi['sedang'])
rule6 = ctrl.Rule(permintaan['sedang'] & persediaan['tinggi'], produksi['tinggi'])
rule7 = ctrl.Rule(permintaan['tinggi'] & persediaan['rendah'], produksi['sedang'])
rule8 = ctrl.Rule(permintaan['tinggi'] & persediaan['sedang'], produksi['tinggi'])
rule9 = ctrl.Rule(permintaan['tinggi'] & persediaan['tinggi'], produksi['tinggi'])
```

# Rule Evaluation

```python
Jumlahproduksi_ctrl = ctrl.ControlSystem([rule1,rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9])
```

# Defuzzifikasi

```python
Jumlahproduksi = ctrl.ControlSystemSimulation(Jumlahproduksi_ctrl)
```

```python
Jumlahproduksi.input['persediaan'] = 800
Jumlahproduksi.input['permintaan'] = 1200
```

```python
Jumlahproduksi.compute()
```

# Hasil Ouput Defuzzifikasi

```python
print(Jumlahproduksi.output['produksi'])
```
