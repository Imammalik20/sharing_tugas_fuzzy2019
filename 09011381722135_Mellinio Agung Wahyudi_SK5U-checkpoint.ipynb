- Fuzzy Rules

```python
rule1  = np.fmin(nilai_nipa_normal, nilai_nips_normal)
ipa_rendah = np.fmin(rule1, ipa_rendah)
rule2  = np.fmin(nilai_nipa_rendah, nilai_nips_normal)
ips_sedang = np.fmin(rule2, ips_sedang)
rule3  = np.fmin(nilai_nipa_rendah, nilai_nips_tinggi)
ips_tinggi = np.fmin(rule2, ips_tinggi)
rule4  = np.fmin(nilai_nipa_rendah, nilai_nips_normal)
ips_sedang = np.fmin(rule4, ips_sedang)
rule5  = np.fmin(nilai_nipa_normal, nilai_nips_tinggi)
ips_tinggi = np.fmin(rule5, ips_tinggi)
rule6  = np.fmin(nilai_nipa_normal, nilai_nips_tinggi)
ips_rendah = np.fmin(rule6, ipa_rendah)
rule7  = np.fmin(nilai_nipa_tinggi, nilai_nips_rendah)
ipa_tinggi = np.fmin(rule7, ipa_tinggi)
rule8  = np.fmin(nilai_nipa_normal, nilai_nips_normal)
ipa_normal = np.fmin(rule8, ipa_normal)
rule9  = np.fmin(nilai_nipa_normal, nilai_nips_rendah)
ipa_normal = np.fmin(rule9, ipa_normal)
```
