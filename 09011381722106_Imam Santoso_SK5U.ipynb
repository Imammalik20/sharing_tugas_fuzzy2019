Dalam operasional kapal dibutuhkan infrastruktur teknologi yang paling terbaru. Kebutuhan infrastruktur
tersebut diantaranya adalah supporting system di kantor dan dikapal. Hal ini di provide oleh fungsi

Sehingga fungsi IT Shipping dapat mengetahui perkiraan barang yang akan di sediakan untuk kebutuhan di

jumlah pengadaan barang yang akan disediakan berikutnya


```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt
```

Input  Persediaan Barang 188 & Permintaan 12 Barang

Output : Jumlah pembelian dibagi lagi menjadi 3 bagian :
- Sedikit (0-97) 
- Sedang (90-170) 
- Banyak (165-243)

### 1. Variabel Input
Variabel input atau faktor yang menentukan besarnya bonus :
- Persediaan

```python
persediaan = ctrl.Antecedent(np.arange(0, 243, 1), 'persediaan')
x_persediaan  = np.arange(0, 243, 1)
```

- Permintaan

```python
permintaan = ctrl.Antecedent(np.arange(4, 20, 1), 'permintaan')
```

### 2. Variabel Ouput
Variabel output ini adalah jumlah bonus :

- Pembelian

```python
pembelian = ctrl.Consequent(np.arange(97, 146, 1), 'pembelian')
```

Fuzzy Rules

## Menghitung Rule Z

- Setelah membandingkan ketiga nilai tersebut, Nilai dari setiap Rule tersebut akan Menjadi nilai PREDIC untuk variable Nilai Hasil Atau Variable output. Nilai PREDIC digunakan untuk mencari Nilai Z dan ketiga nilai tersebut digunakan mencari Nilai total.Menampilkan Jumlah Pembelian dalam Bentuk Grafik

#CATATAN : Untuk Rule 1 sd 5, 7 tidak dihitung karena Nilai PREDIC pada rule tersebut adalah 0, 
karena salah satu dari perbandingan rule bernilai 0 dan tidak akan berpengaruh pada nilai hasil. 
Perhitungan total hanya menggunakan Rule 6 & 9

Rule 6
IF Persediaan Sedang AND Permintaan Banyak THEN Pembelian Banyak

Lalu lihat himpunan Pembelian Banyak
$$
    Pembelian ['Banyak'] = \mu Banyak(x)=\begin{cases}
    (x - 97)/(20 - 97) , & \text{$ 97 \leq x \leq 146 $}.\\
     \end{cases}
 $$


$$
(x-97) / (20-97) = 0.3 \\
(x-97) = 0.3 \\ 
 x = 71 \\
$$
Maka nilai Z$_6$= 71



===============================================================================================================

Rule 9
 IF Persediaan Banyak AND Permintaan Banyak THEN Pembelian Banyak
 
 Lalu lihat himpunan Pembelian Banyak
$$
    Pembelian ['Banyak'] = \mu Banyak(x)=\begin{cases}
    \left.(x - 97)\, \right/(20 - 97) , & \text{$97 \leq x \leq 146$}.\\
    \end{cases}
 $$


$$
(x-97) / (20-97) = 0,3 \\
(x-97) = 0,3 \\ 
 x = 70 \\
$$

Maka nilai Z$_9$= 70

===============================================================================================================



Dari sini kita dapat mencari nilai z, yaitu :

Z =$$ \frac{\alpha predikat_1 \times Z_1 +

{\alpha predikat_1 + \alpha predikat_2 +

\alpha predikat_9 }$$
